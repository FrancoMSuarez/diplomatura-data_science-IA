Graficar en un mismo histograma la distribución de la variable *pf_identity* en global, y en Latinoamérica y el caribe. Repetir para la variable *hf_score*. ¿Visualmente, a qué tipo de distribución corresponde cada variable? ¿Es correcto utilizar todos el conjunto de valores disponibles para esa region en estos gráficos?

```python
display_markdown('### Histograma de la Distribuciones de las variables pf_identity y hf_score')
lac_data = dataset[dataset.region == 'Latin America & the Caribbean']
world_dist = dataset[['pf_identity','hf_score']]
plt.figure(figsize=(10,5))
seaborn.distplot([lac_data.pf_identity], label='pf_identity en Latinoamérica y el Caribe', kde=False, bins = 25)
seaborn.distplot(dataset.pf_identity, label='pf_identity en el Mundo',kde=False, bins = 25)
plt.title('Distribución pf_identity')
plt.xlim(0,10)
plt.legend()
# Histograma hf_score
plt.figure(figsize=(10,5))
seaborn.distplot([lac_data.hf_score], label='hf_score en Latinoamérica y el Caribe', bins=20)
seaborn.distplot(dataset.hf_score, label='hf_score en el Mundo', bins=20)
plt.title('Distribución pf_score')
plt.xlim(2,10)
plt.legend()
```

```python
plt.figure(figsize=(10, 10))
data = lac_data[['hf_score', 'pf_identity']]
seaborn.boxplot(data=data, color='blue')
plt.title('hf_score y pf_identity en el mundo')
plt.xticks(rotation=45)
```

#### Conclusion
Visualmente la variable **hf_score** presenta un distribución Normal.

Visualmente la variable **pf_identity** presenta una distribución Beta.

Usando como base, la información proporcionada por el diagrama de caja. Podemos concluir:
* En el caso de la variable **hf_score** **no** es conveniente utilizar el conjunto completo de valores. Mediante el diagrama de caja podemos observar que esta variable posee outlier.
* En el caso de la variable **pf_identity**, es **correcto** utilizar conjunto de valores, el cual, no presenta outliers.
