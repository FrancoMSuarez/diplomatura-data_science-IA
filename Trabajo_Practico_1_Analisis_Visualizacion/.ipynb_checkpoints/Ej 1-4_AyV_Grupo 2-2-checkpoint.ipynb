### Categorical Analysis

```python
import numpy as np
```

```python
seaborn.lmplot(data=dataset, x='pf_identity', y='pf_score', x_estimator=np.mean, height=7)
```

```python
seaborn.lmplot(data=dataset, x='pf_identity', y='ef_score', x_estimator=np.mean, height=7)
```

**Conclusiones:**
* Hay una relación lineal entre las libertades de identidad y las libertades personales. 
* Las libertades personales son mayores que las libertades de identidad.
* Las libertades económicas tienen cierta relación a las libertades de identidad, pero es baja. 

### <font color=dodgerblue>4.3</font>

### Segmentación

```python
dataset['pf_identity_segment'] = pandas.cut(dataset.pf_identity, 3, labels=["low", "medium", "high"])
dataset['pf_score_segment'] = pandas.cut(dataset.pf_score, 3, labels=["low", "medium", "high"])
dataset[['pf_identity_segment', 'pf_score_segment']][:5]
```

```python
dataset['ef_score_segment'] = pandas.qcut(dataset.ef_score, 5)
dataset[['pf_identity_segment', 'pf_score_segment', 'ef_score_segment']][:5]
dataset.ef_score_segment = dataset.ef_score_segment.apply(lambda x: str(x)).factorize()[0]
dataset[['pf_identity_segment', 'pf_score_segment', 'ef_score_segment']].head()
```

**Conclusion:**
- Se usan las mismas categorizaciones en las libertades personales.
- Se usan categorías de acuerdo a quantiles, cada 20%, dada la distribución de las libertades económicas.

### <font color=dodgerblue>4.4</font>

### Correlación por Segmentos

```python
plt.figure(figsize=(10,6))
seaborn.heatmap(pandas.crosstab(dataset.ef_score_segment, dataset.pf_score_segment))
```

```python
plt.figure(figsize=(10,6))
seaborn.heatmap(pandas.crosstab(dataset.pf_score_segment, dataset.pf_identity_segment))
```

```python
plt.figure(figsize=(10,6))
seaborn.heatmap(pandas.crosstab(dataset.ef_score_segment, dataset.pf_identity_segment))
```
